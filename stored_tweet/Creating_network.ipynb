```python
import networkx as nx
import pandas as pd
import numpy as np
import re

import os
import glob
# use glob to get all the csv files in the folder

path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "*.csv"))

for f in csv_files:
      
    # read the csv file
    df = pd.read_csv(f)
      
    # print the location and filename
    print('Location:', f)
    print('File Name:', f.split("\\")[-1])
      
    # print the content
    print('Content:')
    display(df)
    print()
```

```python
# Read each CSV file into DataFrame
# This creates a list of dataframes
df_list = (pd.read_csv(file) for file in csv_files)

# Concatenate all DataFrames
big_df   = pd.concat(df_list, ignore_index=True)
```

```python
big_df.info()
```

```python
df = big_df.drop(["geo", "retweet_id", "conversation_id", "time", "timezone","retweet_date", "id", "translate", "trans_src", "trans_dest", "near", "source", "user_rt_id","user_rt", "place", "thumbnail", "video", "language", "photos" , "created_at", "cashtags", "urls", "link", "quote_url"], axis=1)
```
