```python
#Importing necessary library
import snscrape.modules.twitter as sntwitter
import pandas as pd
import itertools
from os import mkdir, path
import os
import glob
```

```python
htag=['bonino','+europa','meloni','fratelliditalia','calenda', 'salvini','lega', 'azione', 'letta','partitodemocratico', 'renzi','m5s', 'conte']
```

```python
df_tutti=pd.DataFrame()
for htag in htag:
    df=pd.DataFrame()
    print(htag)
    for j in range(8,10):
        if j==8:
            for i in range (10,31):
                tweet= pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(f'"{htag} since:2022-0{j}-{i} until:2022-0{j}-{i+1} lang:it"').get_items(),100))
                df=df.append(tweet)    
        else:
            for k in range(1, 29):
                tweet= pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(f'"{htag} since:2022-0{j}-{k} until:2022-0{j}-{k+1} lang:it"').get_items(),100))
                df=df.append(tweet)
    df.reset_index()
    df.to_csv(f'{htag}.csv')
    df_tutti=df_tutti.append(df)
                
                    
```

```python
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "*.csv"))
csv_files
```

```python
# Read each CSV file into DataFrame
# This creates a list of dataframes
df_list = (pd.read_csv(file) for file in csv_files)

# Concatenate all DataFrames
big_df   = pd.concat(df_list, ignore_index=True)
big_df=big_df.drop_duplicates('renderedContent', keep='first')
big_df.to_csv('df_tutti_1.csv',index=False)
big_df.info()
```
